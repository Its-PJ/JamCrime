## ***SETUP, CONVERSIONS, AND IMPORTS***

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# from google.colab import files
df = pd.read_csv('seecrime.csv')

df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m-%d')

df['month_year'] = df['DATE'].dt.to_period('M')
```

# ***THIS LINE GRAPH SHOWS THE NUMBER OF VICTIMS PER MONTH***

```python
monthly_victims = df.groupby('month_year')['NUMBER_OF_VICTIMS'].sum()

monthly_victims.index = monthly_victims.index.strftime('%Y-%m')

plt.figure(figsize=(10, 6))
plt.plot(monthly_victims.index, monthly_victims.values, marker='o', linestyle='-', color='#ff6361')
plt.title('Number of Victims per Month in 2023')
plt.xlabel('Month')
plt.ylabel('Number of Victims')
plt.xticks(rotation=45)
plt.grid(True)

plt.ylim(ymin=140)

# Set custom x-axis tick labels
custom_x_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(monthly_victims.index, custom_x_labels, rotation=45)

plt.tight_layout()
plt.show()
```

# ***THESE PIE CHARTS SHOW THE CRIME RATES PER LOCATION***

```python
locations = df['LOCATION'].unique()
```

```python
# Define a list of colors for each crime type (in the same order as they appear in the data)
crime_colors = {
    'Fraud': '#003f5c',
    'Homicide': '#58508d',
    'Burglary': '#bc5090',
    'Assault': '#ff6361',
    'Robbery': '#ffa600'
}

# Iterate through each location
for location in locations:
    # Filter the data for the current location
    location_data = df[df['LOCATION'] == location]

    # Group the data by crime type and calculate the count of each crime
    crime_counts = location_data['CRIME_TYPE'].value_counts()

    # Calculate the total number of crimes for the location
    total_crimes = len(location_data)

    # Create a list of colors in the order of the labels
    colors = [crime_colors.get(crime, 'gray') for crime in crime_counts.index]

    # Create a pie chart for the current location using specific colors
    plt.figure(figsize=(6, 6))
    plt.pie(crime_counts, labels=crime_counts.index, autopct=lambda p: f'{int(p * total_crimes / 100)} ({p:.1f}%)', startangle=140, colors=colors)
    plt.title(f'Crime Rates in {location} for 2023 \n (Total: {total_crimes} crimes) \n')

    # Display the pie chart
    plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
    plt.show()
```
